```python
import numpy as np
import random
```

```python
bias = 1
X = [
    [1, 1, 2],
    [1, 0, 1],
    [0, 1, 1],
    [0, 0, 0]
]
w = [
    random.random(),
    random.random(),
    random.random()
]
```

```python
w
```

```python
def af(num):
    if num > 1:
        return 2
    elif num > 0:
        return 1
    else:
        return 0
```

```python
def noron(instance):
    x0 = instance[0]
    x1 = instance[1]
    y = instance[2]
    output = af(x0*w[0] + x1*w[1] + bias*w[2])
    print(f'Output: {output}')
    print(f'Target: {y}')
    error = y - output
    w[0] = w[0] + (error * x0)
    w[1] = w[1] + (error * x1)
    w[2] = w[2] + (error * bias)
    print(f'Weights {w}')
    return error
```

```python
for j in X:
    error = noron(j)
    print(error)
```

```python
for i in range(100):
    print('epoch: ' + str(i))
    for j in X:
        error = noron(j)
        print(error)
```
