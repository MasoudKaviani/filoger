```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
```

```python
data = pd.read_csv('shuttle.csv', header=None, names=['LastLoginDays', 'NumberOfLoginLastWeek', 'NumberOfLikesLastWeek', 'IIE', 'NLS', 'LKK', 'LAI', 'LAK', 'ODK', 'OKO'])
data.head()
```

```python
IF = IsolationForest(contamination=0.01)
y_pred = IF.fit_predict(data.values)
```

```python
np.unique(y_pred, return_counts=True)
```

```python
data['Anomaly'] = y_pred
```
