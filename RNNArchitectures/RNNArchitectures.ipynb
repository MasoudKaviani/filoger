```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import TimeDistributed, Dense, LSTM, GRU, SimpleRNN
```

---
One to One

```python
length = 10
seq = np.array([ i/float(length) for i in range(length)])
X = seq.reshape(len(seq), 1, 1)
y = seq.reshape(len(seq), 1)
n_neuros = length
n_batch = length
n_epoch = 100
```

```python
X
```

```python
model  = Sequential()
model.add(GRU(n_neuros, input_shape=(1, 1)))
model.add(Dense(1))
```

```python
model.summary()
```

```python
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model.fit(X, y, epochs=n_epoch, batch_size=n_batch)
```

Many to One

```python
X = seq.reshape(1, length, 1)
y = seq.reshape(1, length)
```

```python
X
```

```python
y
```

```python
model  = Sequential()
model.add(GRU(n_neuros, input_shape=(length, 1)))
model.add(Dense(length))
```

```python
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model.summary()
```

```python
model.fit(X, y, epochs=n_epoch, batch_size=n_batch)
```

```python
model.predict(X, batch_size=n_batch)
```

---
Many 2 Many

```python
X = seq.reshape(1, length, 1)
y = seq.reshape(1, length, 1)
```

```python
X
```

```python
y
```

```python
model  = Sequential()
model.add(GRU(n_neuros, input_shape=(length, 1), return_sequences=True))
model.add(TimeDistributed(Dense(length)))
```

```python
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model.summary()
```

```python
model.fit(X, y, epochs=n_epoch, batch_size=n_batch)
```

```python
model.predict(X, batch_size=n_batch)
```
